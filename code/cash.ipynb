# Clean Data

```python
def from_dt_to_mon(timestemp):
    return datetime.fromisoformat(timestemp).strftime("%B")[:3]

def from_dt_to_yr(timestemp):
    return datetime.fromisoformat(timestemp).strftime("%Y")
```

```python
df_cash_req["month"] = df_cash_req["created_at"].apply(from_dt_to_mon)
```

```python
df_cash_req["year"] = df_cash_req["created_at"].apply(from_dt_to_yr)
```

Status of the CR.
Possibles values are : 
- approved : CR is a 'regular' one (= without fees) and was approved either automatically or manually. Funds will be sent aprox. 7 days after the creation
- money_sent : We transferred the fund to the customer account. Will change to active once we detect that the user received the funds (using user's bank history)
- rejected : The CR needed a manual review and was rejected
- pending : The CR is pending a manual review from an analyst
- transaction_declined : We failed to send the funds to the customer
- waiting_user_confirmation : The user needs to confirm in-app that he want the CR (for legal reasons)
- direct_debit_rejected : Our last attempt of SEPA direct debit to charge the customer was rejected 
- canceled : The user didn't confirm the cash request in-app, we automatically canceled it
- direct_debit_sent : We sent/scheduled a SEPA direct debit to charge the customer account. The result of this debit is not yet confirmed
- waiting_reimbursement : We were not able to estimate a date of reimbursement, the user needs to choose one in the app.
- active : Funds were received on the customer account.
- money_back : The CR was successfully reimbursed.

```python
df_cash_req_userid_filled = df_cash_req.copy()

# fill user ID with Deleted Account ID, when empty
df_cash_req_userid_filled["user_id"] = df_cash_req_userid_filled["user_id"].fillna(df_cash_req_userid_filled["deleted_account_id"])
df_cash_req_userid_filled
```

```python
df_cash_req_userid_filled.groupby(["year","month"])["user_id"].nunique()
```

```python
# group for user id and get combinations for their activity
user_activity = df_cash_req_userid_filled.groupby('user_id').agg({'year': pd.Series.nunique, 'month': pd.Series.nunique})

# filter users for activity in more than one year and/or month
multiple_months_or_years = user_activity[(user_activity['year'] > 1) | (user_activity['month'] > 1)]

multiple_months_or_years
```

```python
month_to_number = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
                   'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

df_cash_req_userid_filled['date_key'] = pd.to_datetime(df_cash_req_userid_filled['year'].astype(str) + '-' + df_cash_req_userid_filled['month'].map(month_to_number).astype(str) + '-01')
df_cash_req_userid_filled
# Sortiere den DataFrame nach dem neuen Schlssel
df_cash_req_sorted = df_cash_req_userid_filled.sort_values(by=['user_id', 'date_key'])

df_cash_req_sorted
```

```python
# define cohort for user regarding to the key_date (first Cash Request)
cohort = df_cash_req_sorted.groupby('user_id').first().reset_index()

# Print the cohort in the new column
df_cash_req_sorted['cohort'] = df_cash_req_sorted['user_id'].map(cohort.set_index('user_id')['date_key'])

df_cash_req_sorted
```

```python
user_per_cohort = df_cash_req_sorted.groupby("cohort")["user_id"].nunique()
user_per_cohort
```

```python
usage_per_cohort = df_cash_req_sorted["cohort"].value_counts()
usage_per_cohort
```

```python
average_usage_per_user = usage_per_cohort / user_per_cohort
average_usage_per_user
```

```python
cohort_table = pd.concat([user_per_cohort, usage_per_cohort, average_usage_per_user], axis=1)
cohort_table.columns = ['Users', 'Frequency of usage', 'Average usage per user']
cohort_table
```

```python
sns.barplot(data=cohort_table, y = "cohort", x = "Users", palette="Blues")
#plt.rcParams.update({'font.size':18})
plt.title('Users per Cohort')
plt.show()
```

```python
sns.barplot(data=cohort_table, y = "cohort", x = "Frequency of usage", palette="Blues")
#plt.rcParams.update({'font.size':18})
plt.title('Usage per cohort')
plt.show()
```

```python
sns.barplot(data=cohort_table, y = "cohort", x = "Average usage per user", palette="Blues")
#plt.rcParams.update({'font.size':18})
plt.title('User Activity')
plt.show()
```
